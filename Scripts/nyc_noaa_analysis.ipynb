# DATA WRANGLING 

```python
#Secure all items  that correspond to the TAVG as avg_temp

avg_temps = [item for item in d['results'] if item['datatype']=='TAVG']
```

```python
#Getting the dates from the avg_temp

dates_temp = [item['date'] for item in avg_temps]
```

```python
#Get the temperature from all the avg_temp

temps = [item['value'] for  item in avg_temps]
```

```python
#Checking the results in temps

temps
```

```python
#Adding the results in an empty DataFrame where I will be adding new columns

df_temp = pd.DataFrame()
```

```python
# Adding new columms into df_temp in this case date and avg_temp

df_temp['date']= [datetime.strptime(d,"%Y-%m-%dT%H:%M:%S") for d in dates_temp]
df_temp['avgTemp']=[float(v)/10.0 for v in temps]
```
